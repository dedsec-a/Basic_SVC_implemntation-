```python
# We are Going to use the tips dataset for this particular Predection 
# importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as ptl
import seaborn as sns

```

```python
df = sns.load_dataset('tips')
df.head()
```

# PROBLEM STATMENT 
 In This particular case we are going to predict the total bill of  based on the other features such as tip , sex, smoker , day , time 
 and Size

```python
# Checking diffrent types of cataogies in the given catagorey dataset
print(df['sex'].value_counts())
print(df['smoker'].value_counts())
print(df['day'].value_counts())
print(df['time'].value_counts())
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
X = df.drop('total_bill' , axis= 1)
y = df['total_bill']
l1 = LabelEncoder()
l2 = LabelEncoder()
l3 = LabelEncoder()
```

```python
print(X)
print(y)
```

```python
# Spliting the data into Traning and Testing split
x_train , x_test , y_train , y_test = train_test_split(X, y , test_size= 0.25 ,random_state= 42)
```

```python
# We Can do Encoding only After the Train Test Split to prevent data-encoding
x_train.head()
```

```python
x_train['sex'] = l1.fit_transform(x_train['sex'])
x_train['smoker'] = l2.fit_transform(x_train['smoker'])
x_train['time'] = l3.fit_transform(x_train['time'])
```

```python
x_train.head()
```

```python
x_test['sex'] = l1.fit_transform(x_test['sex'])
x_test['smoker'] = l2.fit_transform(x_test['smoker'])
x_test['time'] = l3.fit_transform(x_test['time'])
```

```python
# OneHotEncoding --> Column transformer 
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer(transformers=[('onehot', OneHotEncoder(drop= 'first'),[3])] , remainder='passthrough')
on = OneHotEncoder()
```

```python
x_train = ct.fit_transform(x_train)
```

```python
x_train
```

```python

x_test = ct.transform(x_test)
```

```python
# importing the support vector machine 
from sklearn.svm import SVR 
svr = SVR()
svr.fit(x_train , y_train)
```

```python
y_preds = svr.predict(x_test)
```

```python
from sklearn.metrics import r2_score , mean_absolute_error
print(r2_score(y_test , y_preds))
print(mean_absolute_error(y_test , y_preds))
```

```python
from sklearn.model_selection import GridSearchCV
praram_grid =  {
    'C' : [0.1 , 1 , 10 , ],
    'gamma' : [1, 0.1 , 0.01 , 0.001 , 0.0001],
    'kernel' : ['rbf'] 
}
```

```python
grid = GridSearchCV(SVR() , praram_grid , refit=True , verbose= 3 )
grid.fit(x_train , y_train )
```

```python
print(grid.best_params_)
grid.best_estimator_
```

```python
grid_pred = grid.predict(x_test)
```

```python
print(r2_score(y_test , grid_pred))
print(mean_absolute_error(y_test , grid_pred))
```

We can see that the model has improved  as the r2 Score has gone up and the Mean Absolute Error has Gone Down
